There were an estimated 18.1 million new cases of cancer worldwide in 2020. Breast cancer and lung cancer were the most common types accounting for 12.5% and 12.2% of cases respectively. Each year in Ireland, more than 3,500 women and approximately 35 men are diagnosed with breast cancer.  It is more common in women over 50. Breast cancer is treated with surgery, radiotherapy, chemotherapy, hormone therapy and targeted therapies, depending on the type. Finding cancer early saves lives and can mean that the cancer is easier to treat. Less than 1 in 100 women screened are diagnosed with cancer, and these women have an good chance of successful treatment. Breast screening is a very effective way of spotting breast cancer early. Cancer is a disease that is caused by groth of cells in an uncontrolled maner in the body. These rapidly dividing cells form masses in the breast called tumors. Tumors are divided into two groups as benign and malignant. Malignant tumors grow and damage adjacent healthy tissue. Harmful cells inside the tumor can also spread to different organs of the body. This is known as metastasis.

There is a good chance of recovery from breast cancer if it is identified an early stage and treated, Therefore many studies have been carried out on its early diagnosis. Quite alot of research has gone into imaging and image analysis. A number of different medical imaging techniques are commonly used for diagnosis, more so than any other testing method. Medical imaging modalities used for breast cancer detection include histopathology (hp) images, breast X-ray images (mammography), sonograms (ultrasound imaging), and magnetic resonance imaging (MRI) are broadly used for diagnosis, with the mammogram being the most common method used. If something suspicious is found following this imaging a biopsy is often carried out. A common method of breast cancer biopsy is a Fine Needle Aspiration (FNA). During a FNA, a small amount of breast tissue or fluid is removed from a suspicious area with a thin, hollow needle and checked for cancer cells.

### The Wisconsin Diagnostic Bresat Cancer Dataset

The Wisconsin Diagnostic Bresat Cancer Dataset (WDBC) is seen by many as a benchmark dataset for training classification algorithms and is often referred to in these studies. The WDBC dataset is often used because of the relativley large number of inatances in the dataset is also relativley noise-free with fewoutliers or missing values.

The WDBC is a classification dataset, which records the measurements for breast cancer cases. This dataset came from the work of Dr. William H. Wolberg and his team at the departments of Surgery and Human Oncology at the University of Wisconsin-Madison. (Machine Learning for Cancer Diagnosis and Prognosis, Dec 2022) [4] He and his team had a desire to diagnose breast masses accurately based solely on a FNA. They used fluid samples, taken from patients with solid breast masses and computer program called Xcyt to analyse the images. They identified nine visually assessed characteristics of an FNA sample which they considered relevant to diagnosis.  In collaboration with Prof. Mangasarian and two of his graduate students, Rudy Setiono and Kristin Bennett, a classifier was constructed using the multisurface method (MSM) of pattern separation on these nine features that successfully diagnosed 97% of new cases. The resulting data set is well-known as the Wisconsin Breast Cancer Data. It was donated in 1995 be W. Nick Street of the Computer Sciences Dept at the University of Wisconsin. [2]

#### Dataset detailed information:

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image. A few of the images can be found at [Web Link]


Separating plane described above was obtained using Multisurface Method-Tree (MSM-T) [K. P. Bennett, "Decision Tree Construction Via Linear Programming." Proceedings of the 4th Midwest Artificial Intelligence and Cognitive Science Society, pp. 97-101, 1992], a classification method which uses linear programming to construct a decision tree. Relevant features were selected using an exhaustive search in the space of 1-4 features and 1-3 separating planes.

The actual linear program used to obtain the separating plane in the 3-dimensional space is that described in: [K. P. Bennett and O. L. Mangasarian: "Robust Linear Programming Discrimination of Two Linearly Inseparable Sets", Optimization Methods and Software 1, 1992, 23-34].
[2](uci breast cancer wisconsin, Dec 2022)

#### Database Attribute Information:

1) ID number
2) Diagnosis (M = malignant, B = benign)
3-32)

Ten real-valued features are computed for each cell nucleus:

a) radius (mean of distances from center to points on the perimeter)
b) texture (standard deviation of gray-scale values)
c) perimeter
d) area
e) smoothness (local variation in radius lengths)
f) compactness (perimeter^2 / area - 1.0)
g) concavity (severity of concave portions of the contour)
h) concave points (number of concave portions of the contour)
i) symmetry
j) fractal dimension ("coastline approximation" - 1)

[2] (uci breast cancer wisconsin, Dec 2022) https://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+(diagnostic)

### Feature Selection
Feature selection is a key step in the project. I propose to select the best attributes to test by correlation-based feature selection (CFS), That is corrolation of the feature to the diagnosis. We can then evaluate based on their corrolation. The aim is to reduce the number of input variables by removing those that are precieved to have little corrolation to the diagnosis. In simple terms, if we put alot of rubbish into out algorithms we will get rubbish out. Apart from choosing the right model for our data, we need to choose the right data to put in our model. [16] In the case of the WDBC for example, the case ID number may confuse the algorithm, It may find some corrolation in the ID number that may skew the results so we can remove that column with some confidence we will not loose important features. It is important to note, while I am removing the case ID from the DF at this point, the information has not been discarded entirley. It may be the case that we revisit and include it at a later date. Case ID may for example relate to paitents geographical area which in turn may have some bearing on the likley diagnosis.

Feature selection methods can generally be described as either supervised or unsupervised. Supervised selection methods include information of a given classe in the selection, while unsupervised methods can be used for selection without class labels.

### Litreature review

Many studies have been carried out on the WDBC with a view to classifying samples correctly based on the data available. Some of the most common and successful techniques are:
Decision Tree (DT),
Random forest,
Artificial Neural Networks (ANN), 
Support Vector Machines (SVM), 
Naive Bayes multinomial classifier (NBM),
K-nearest neighbors (KNN)
Logistic Regression

Decision Tree 
Decision Trees are a type of Supervised Machine Learning. You explain what the input is and what the corresponding output is in the training data. The data is repeatedly split according to a certain parameter. The tree will begin to grow based on two enteties called decision nodes and leaves. The resultant tree can be seen as a piecewise constant approximation.

Random forest classifier.
A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

Artificial Neural Network.
A neural network consists of units (neurons), arranged in layers, which convert an input vector into some output. Each unit takes an input, applies a (often nonlinear) function to it and then passes the output on to the next layer.

Support Vector Machines 
Support vector machines (SVMs) are a set of supervised learning methods used for classification, regression and outliers detection. Developed at AT&T Bell Laboratories in 1997 SVMs are one of the most robust prediction methods, being based on statistical learning frameworks.

Bayesian Classifier
Naive Bayes methods are a set of supervised learning algorithms based on applying Bayes’ theorem with the assumption of conditional independence between every pair of features. This method is called “naive” because the assumption that different attributes are independent of each other.

K-nearest neighbors
The principle behind nearest neighbor methods is to find a predefined number of training samples closest in distance to the new point, and predict the label from these. The number of samples can be a user-defined constant (k-nearest neighbor learning), or vary based on the local density of points (radius-based neighbor learning)

Logistic regression
Despite its name, it is implemented as a linear model for classification rather than regression. The logistic regression is also known in the sklearn literature as logit regression, maximum-entropy classification (MaxEnt) or the log-linear classifier.
