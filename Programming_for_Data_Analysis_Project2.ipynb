#### Dataset detailed information:

Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei.

Separating plane described above was obtained using Multi-surface Method-Tree (MSM-T) [K. P. Bennett, “Decision Tree Construction Via Linear Programming.” Proceedings of the 4th Midwest Artificial Intelligence and Cognitive Science Society, pp. 97–101, 1992], a classification method which uses linear programming to construct a decision tree. Relevant features were selected using an exhaustive search in the space of 1–4 features and 1–3 separating planes.

The actual linear program used to get the separating plane in the 3-dimensional space is that described in: [K. P. Bennett and O. L. Mangasarian: “Robust Linear Programming Discrimination of Two Linearly Inseparable Sets”, Optimization Methods and Software 1, 1992, 23–34].

[2] (uci breast cancer wisconsin, Dec 2022)

#### Database Attribute Information:

1) ID number
2) Diagnosis (M = malignant, B = benign)
3-32)

Ten real-valued features are computed for each cell nucleus:

a) radius (mean of distances from center to points on the perimeter)
b) texture (standard deviation of gray-scale values)
c) perimeter
d) area
e) smoothness (local variation in radius lengths)
f) compactness (perimeter^2 / area - 1.0)
g) concavity (severity of concave portions of the contour)
h) concave points (number of concave portions of the contour)
i) symmetry
j) fractal dimension ("coastline approximation" - 1)

##### Random forest classifier.
A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

##### Artificial Neural Network.
A neural network consists of units (neurons), arranged in layers, which convert an input vector into some output. Each unit takes an input, applies a (often nonlinear) function to it and then passes the output on to the next layer.

##### Support Vector Machines

##### Bayesian Classifier

##### K-nearest neighbors
The principle behind nearest neighbor methods is to find a predefined number of training samples closest in distance to the new point, and predict the label from these. #### Fixing Multicollinearity and final Feature Selection.

It is important to address features that show multicollinearity during data processing. The information provided by the collinear variables is somewhat redundant, so removal of the appropriate variables will reduce inputs, hopefully without significant impact on the results.

Using VIF — Variance Inflation Factor will help select these variables. VIF calculates how much the variance of a coefficient is inflated because of its linear dependencies with other predictors.

The scatterplot and correlation matrix used already show only bivariate relationships, where as VIF shows the relationship within a group of variables.

To calculate VIF we take one feature and check regerssion against all the others, then repeat again for each feature. With the help of sklearn LinearRegression the function below calculates the VIF. 

[20] (Variance Inflation Factor (VIF), CFI Team, Oct 2022)

And Code from [19]  (Statistics in Python — Collinearity and Multicollinearity, towardsdatascience, Wei-Meng Lee, Oct 2021)

This information helps with feature selection. 

```python
#Feature selection

import pandas as pd
from sklearn.linear_model import LinearRegression
def calculate_vif(df, features):    
    vif, tolerance = {}, {}
    for feature in features:
        X = [f for f in features if f != feature]        
        X, y = df[X], df[feature]
        r2 = LinearRegression().fit(X, y).score(X, y)                
        tolerance[feature] = 1 - r2
        vif[feature] = 1/(tolerance[feature])
    return pd.DataFrame({'VIF': vif, 'Tolerance': tolerance})
```

```python
#calculate_vif(df=df_mean_features, features=df_mean_features.columns[:10])
calculate_vif(df=df_mean_features, features=['radius_mean', 
                                             'texture_mean',
                                             'perimeter_mean', 
                                             'area_mean', 
                                             'smoothness_mean', 
                                             'compactness_mean', 
                                             'concavity_mean', 
                                             'concave points_mean', 
                                             'symmetry_mean',
                                             'fractal_dimension_mean'])
```

#### What does the VIF factor tell us?

A low value equates to low correlation. 

VIF = 1: A VIF factor of 1 shows the variable is not correlated to the remaining ones. Multicollinearity does not exist.

VIF = 1--5: Somewhat correlated

VIF = 5+: Highly correlated

```python
#Funtion to train and score models
def train_score(models, X_train, X_test, y_train, y_test):
  # Set random seed
  np.random.seed(0)
  # Make an empty dictionary for model scores
  scores = {}  
  # Loop through models
  for name, model in models.items():
      #Fit the model to the data
      model.fit(X_train, y_train)
#Evaluate the model and append its score to the scores dictionary
      scores[name] = model.score(X_test, y_test)
  return scores
```

```python
scores = train_score(models = models,
                   X_train = X_train,
                   X_test = X_test,
                   y_train = y_train,
                   y_test = y_test)

print(scores)
```

```python
#Model comparison
model_compare = pd.DataFrame(scores, index=["Accuracy"])
model_compare.T.plot.bar();
plt.title('corr_features')
plt.savefig('data/corr_features.png')
```

<h1><center>Extending the dataset</center></h1>

To continue improving a machine learning model we could use more data. Building robust machine learning models requires large amounts of data. This data could be real world, or there is the option to extend or replicate the existing dataset by synthesis. Important factors to consider would be distribution types, corrolation between colums, properties of each column including standard deviation, mean, min, max etc.

There are some issues using real world data. As well as being difficult and time consuming to collect, real world data often has errors or contains a bias that may affect the quality of a ML model built on it. It is often difficult to obtain relevant data on the large scale required for training and testing a predictive model, especially on a project like this based on medical records. Datasets containig sensitive presonal information needs to be handled sensitivley and securley and may require special permissions. Synthetic data when generated correctly can fill the gaps by supplementing real world data on a large scale quickly.

Types of synthetic data include Fully synthetic, Partially synthetic and a Hybrid approach. I created a dataset based on the featured and parameters of the WDBC dataset, but it retains none of the original data so is considered fully synthetic. A partially synthetic dataset replaces the values of certain features with synthetic values while still retaining some of the real data. This approach is sometimes used to fill in gaps in the original data. Hybrid data combines both real and synthetic data. It pairs random records from a real dataset with close synthetic records. Hybrid data offers benefits but requires alot of processing power, memory and programming skills.
